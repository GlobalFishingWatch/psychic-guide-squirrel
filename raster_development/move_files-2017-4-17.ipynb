```python
from os import listdir
from os.path import isfile, join
import os 
import subprocess

mypath = 'data/dailytables/fishing_vessel_effort_iso3_gears/tifs'
files = [f for f in listdir(mypath) if isfile(join(mypath, f))]

dirs = [d for d in os.listdir(mypath) if os.path.isdir(join(mypath, d))]
```

```python
# now do the dangerous thing and delete all the files

import subprocess

command = '''earthengine rm -r projects/globalfishingwatch/'''
commands = []
for d in dirs[1:]:
    if d != "start":
        commands.append(command + d)


for command in commands:
    p = subprocess.Popen(command, stdout=subprocess.PIPE, shell=True)

    (output, err) = p.communicate()  

    #This makes the wait possible
    p_status = p.wait()

    #This will give you the output of the command being executed
    print command
```

```python
# delete all the old files in gcs

for d in dirs:
    command = '''gsutil -m rm -f gs://david-scratch/iso3_imagecollections/{iso3}/*'''.format(iso3=d)
    #os.system(command)
    # only if you want to remove everything...
```

```python
# upload all the files to gcs

commands = []

for d in dirs:
    if d != "WLD" and d != "start":
        command = 'gsutil -m cp data/dailytables/fishing_vessel_effort_iso3_gears/tifs/{iso3}/* \
        gs://david-scratch/iso3_imagecollections/{iso3}/'.format(iso3=d)
        commands.append(command)
        os.system(command)

```

```python
# now upload all the files to earth engine

for d in dirs:
    if d != "start":
        command = '''python ee-density-upload-iso3-20170412.py gs://david-scratch/iso3_imagecollections/{iso3}\
 projects/globalfishingwatch/{iso3}'''.format(iso3=d)
        print command
        os.system(command)
```
