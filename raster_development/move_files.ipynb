```python
# This is to try a new version of earth engine collections
# each country will be its own collection. This notebook is to move
# the images from the current locations and move them to a seperate folder
# in GCS and then create an image collection for each one.
```

```python
from os import listdir
from os.path import isfile, join
mypath = 'data/dailytables/fishing_vessel_effort_iso3_gears/tifs'
files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
iso3s = []
for f in files:
    iso3 = f.split("-")[0]
    if iso3 not in iso3s:
        iso3s.append(iso3)
```

```python
import os
for iso3 in iso3s:
    if iso3 !="Unknown":
        iso3_to = iso3
    else:
        iso3_to = "UKN"
    command = 'gsutil -m cp gs://david-scratch/iso3_gears/{iso3}* gs://david-scratch/iso3_imagecollections/{iso3_to}/'.format(iso3=iso3,
          iso3_to = iso3_to                                                                                                                   )
    print command
    os.system(command)
```

```python
for iso3 in iso3s:
    if iso3 =="Unknown":
        iso3 = "UKN"
    command = '''python ee-density-upload-iso3.py gs://david-scratch/iso3_imagecollections/{iso3} projects/globalfishingwatch/{iso3}'''.format(iso3=iso3)
    print command
```

```python
import subprocess
for iso3 in iso3s:
    command = '''earthengine ls projects/globalfishingwatch/{iso3}'''.format(iso3=iso3).split(" ")

    p = subprocess.Popen(command, stdout=subprocess.PIPE, 
                                        stderr=subprocess.PIPE)
    out, err = p.communicate()
    print iso3, len(out), out.split("\n")[-1]
    if len(out)==0:
        command =  'export PYTHONPATH=/usr/local/lib/python2.7/site-packages:$PYTHONPATH'
        command = '''python ee-density-upload-iso3.py gs://david-scratch/iso3_imagecollections/{iso3} projects/globalfishingwatch/{iso3}'''.format(iso3=iso3)
        p = subprocess.Popen(command.split(" "), stdout=subprocess.PIPE, 
                                            stderr=subprocess.PIPE) 
        out, err = p.communicate()
        print out
        
```
